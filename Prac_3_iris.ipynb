```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
df = pd.DataFrame(iris.data)
```

```python
df
```

```python
df = df.rename(columns={0:'SepalLengthCm', 1:'SepalWidthCm', 2:'PetalLengthCm', 3:'PetalWidthCm'})
```

```python
df
```

```python
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
df
```

```python
setosa = df[df['species'] == 'setosa'].iloc[:, :-1]
```

```python
setosa.mean()
```

```python
setosa.std()
```

```python
setosa.quantile([0.25, 0.50, 0.75])
```
